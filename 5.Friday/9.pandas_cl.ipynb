```python
import pandas as pd
```

```python
df = pd.read_csv("csvmichelin.csv", index_col="id")
df.head()
```

```python
real = ["Spain", "Real Madrid", 15]
milan = ["Italy", "AC Milan", 7]
bayern = ["Germany", "Bayern Munchen", 6]
```

```python
teams_dict = {
    "country":[real[0], milan[0], bayern[0]],
    "team":[real[1], milan[1], bayern[1]],
    "trophies":[real[2], milan[2], bayern[2]]
}
teams_dict
```

```python
pd.DataFrame(teams_dict)
```

```python
big__guns_df = pd.DataFrame(teams_dict)
```

```python
list_of_dfs_to_concat = [big__guns_df, df]
pd.concat(list_of_dfs_to_concat).head(6)
```

```python
list_of_dfs_to_concat = [big__guns_df, df]
pd.concat(list_of_dfs_to_concat, ignore_index=True).head(5)
```

```python
list_of_dfs_to_concat = [df, big__guns_df]
merged_df = pd.concat(list_of_dfs_to_concat, ignore_index=True)
merged_df
# merged_df.head(7)
```

```python
# merged_df.sort_values(["trophies", "team", "country"], ascending = [False, True, True])
merged_df.sort_values(["trophies", "team", "country"], ascending = [False, True, True], inplace=True)
merged_df
```

```python
merged_df.reset_index(inplace=True)
merged_df
```
